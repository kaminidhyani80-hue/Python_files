**Exercise 1:** Wrap a division `a/b` in try/except to catch `ZeroDivisionError` and print a helpful message.

**Exercise 2:** Convert the string `'42'` to int inside try/except/else/finally; print success in else, 'done' in finally.

```python
a = '42'
try:
    a = int(a)
except :
    print("Erro occured")
else:
    print("Success")
finally:
    print("Done")
```

**Exercise 3:** Open a missing file path and catch `FileNotFoundError` gracefully; print a friendly message.

```python
try:
    with open("abc.txt","r") as f:
        print(f.read())
except FileNotFoundError as e:
    print("The file you are tyring to access is not present")    
```

**Exercise 4:** Use `raise` to throw a `ValueError` with a custom message if a number is negative; test with `-1` and catch it.

```python
try:
    a = int(input("Enter the number"))
    if a<0:
        raise ValueError("Number cannot be negetive")
except ValueError as e:
    print(e)
```

**Exercise 5:** Handle multiple exceptions (`ValueError` and `TypeError`) when converting `x` to int for `x in ['15','oops',None]`.

```python
x = ['15','oops',None]
try:
    for i in x:
        int(i)
except ValueError as e:
    print("Value Error")
except TypeError as e:
    print("Type Error")
```

**Exercise 6:** Define a custom exception class `TooSmallError`. Write a function that raises it if input < 10; test with 7 and handle it.

```python
class TooSmallError(Exception):
    pass
try:
    a = int(input("Enter the number"))
    if a<10:
        raise TooSmallError("The number you have entered is too small")
except TooSmallError as e:
    print(e)
```

**Exercise 7:** Read an integer from a string; if conversion succeeds (else branch), print its square; otherwise print 'bad input'. Use `s='16'`.

```python
s = '16'
try:
    s = int(s)
except :
    print("bad input")
else:
    print(s**2)
finally:
    print("conversion succeeds")
```

**Exercise 5:** From a list `['red','green','blue','yellow']`, randomly shuffle the list and print it. Then sample 2 unique colors.

```python
colors = ['red','green','blue','yellow']
random.shuffle(colours)
print(colors)
print(random.choices(colors,k=2))
print(random.sample(colors,2))
```

**Exercise 6:** Simulate 1000 coin flips (`H`/`T`) and print the counts of each.

```python
results = [random.choice(["H","T"])for i in range(1000)]
Heads = results.count("H")
Tails = results.count("T")
print("Heads",Heads)
print("Tails",Tails)
```

**Exercise 9:** Parse `'2025-09-11 08:30:00'` to a `datetime`, add 90 minutes, and print the result in `YYYY-MM-DD HH:MM` format.

```python
from datetime import datetime,timedelta
dte = datetime.strptime('2025-09-11 08:30:00',"%Y-%m-%d %H:%M:%S")
print(dte)
future = dte+timedelta(minutes=90)
print(future)
```

**Exercise 10:** Create a small schedule: start at today's date and print the next 5 day names (Mon..Sun) using `datetime`.

```python
from datetime import datetime, timedelta

# Get today's date
today = datetime.today()

# Loop for next 5 days
for i in range(1,6):
    day = today + timedelta(days=i)
    print(day.strftime("%A"))
```
