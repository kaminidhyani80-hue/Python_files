Here are your test scores:[45,78,90,66].Find total marks ,higest,lowest,and average

```python
test_scores = [45,78,90,66]
print(f"Total marks is{sum(test_scores)}")
print(f"Heighest marks is{max(test_scores)}")
print(f"Lowest marks is{min(test_scores)}")
print(f"Average marks is{sum(test_scores)/len(test_scores)}")
```

If a student scores above 90, you say ‘Excellent’. Else if above 60, say ‘Good’. Else say ‘Try again!’”

```python
scores = int(input("Enter the scores"))
if(scores>90):
    print("Excellent")
elif(scores>60):
    print("Good")
else:
    print("Try again")
```

If your total bill crosses ₹500, you get a 10% discount.

```python
bill = int(input("Enter the bill amount"))
if(bill>500):
    bill = (bill-0.1*bill)
    print(bill)
else:
    print(bill)
```
